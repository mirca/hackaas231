```julia
function juliaset{T <: Real}(f::Function, x::Array{T, 1}, y::Array{T, 1}, n=200)
    image = 2*ones(length(x), length(y))
    for i in 1:length(x)
        for j in 1:length(y)
            z = x[i] + 1im*y[j]
            for k = 1:n
                z = f(z)
                if abs(z) > 2.0
                    image[i, j] = k
                    break
                end
            end
        end
    end
    return image
end
```

```julia
x = [linspace(-2, 2, 300);]
y = [linspace(-2, 2, 400);];
```

```julia
# Galaxy Julia Set
f(z) = z.^2 - 0.75 - 0.2im
image = juliaset(f, x, y, 200);
heatmap(sqrt.(image'))
```

```julia
# GIF
anim = @animate for i in range(4, 4, 200)
    image = juliaset(f, x, y, i);
    heatmap(sqrt.(image'))
end
gif(anim, "galaxy_julia_set.gif", fps=15)
```
